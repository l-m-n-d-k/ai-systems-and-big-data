## Task 3.3.2

```python
import pandas as pd
from sklearn import preprocessing
```

```python
url = 'https://raw.githubusercontent.com/akmand/datasets/master/iris.csv'
df = pd.read_csv(url)
```

```python
minmax=preprocessing.MinMaxScaler(feature_range=(0, 1))
zscaler=preprocessing.StandardScaler()
df[["sepal_length_cm"]]=minmax.fit_transform(df[["sepal_length_cm"]])
df[["sepal_width_cm"]]=zscaler.fit_transform(df[["sepal_width_cm"]])
```

```python
print(df)
```
