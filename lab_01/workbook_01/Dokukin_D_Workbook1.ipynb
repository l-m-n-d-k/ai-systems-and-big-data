## Task 4.3.2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x**2)) / np.abs(1 - np.sin(x)**3) + np.log(2 * np.abs(x))

x_vals = np.linspace(1, 10, 10)
y_vals = f(x_vals)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, marker='o', label='f(x)', color='blue')
plt.title('Function values for x in [1, 10]')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x_vals, y_vals, color='red', s=50, zorder=5, label='f(x)')
plt.title('Function values for x in [1, 10]')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

## Task 4.3.3

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simpson, trapezoid
```

```python
def f(x):
    return np.abs(np.cos(x * np.exp(np.abs(np.cos(x)) + np.log(np.abs(x + 1)))))

x = np.arange(0, 11, 1)
y = f(x)
```

```python
plt.figure(figsize=(8, 5))
# plt.plot(x, y, 'o-', label=r'$f(x) = \left|\cos\left(x e^{|\cos x| + \ln|x+1|}\right)\right|$')
plt.plot(x, y, 'o-', label=r'$f(x)')
plt.title('Graph of f(x) on [0, 10] with step h = 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(x)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
area_trapz = trapezoid(y, x)
area_simps = simpson(y, x)

print(f"Trapezoidal rule (trapz): {area_trapz:.6f}")
print(f"Simpson's rule (simps):   {area_simps:.6f}")
```

## Task 4.3.4

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tickers = ["AAPL", "MSFT", "GOOGL"]
start_date = "2021-01-01"
end_date = "2021-12-31"

data = yf.download(tickers, start=start_date, end=end_date)

close_prices = data['Close']

plt.figure(figsize=(14, 7))
for ticker in tickers:
    plt.plot(close_prices.index, close_prices[ticker], label=ticker)

plt.title('Динамика стоимости акций в 2021 году', fontsize=16)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Цена закрытия, USD', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
